```python
filename = 'puzzle_three.txt'
with open(filename) as f:
    data = f.readlines()
wire_1 = data[0].rstrip('\n').split(',')
wire_2 = data[1].split(',')
```

# Test Cases and function definations

```python
# test_wire2 = ['U7', 'R6', 'D4', 'L4']
# test_wire1 = ['R8', 'U5', 'L5', 'D3']

# test_wire1 = ['R75', 'D30', 'R83', 'U83', 'L12', 'D49', 'R71', 'U7', 'L72']
# test_wire2 = ['U62', 'R66', 'U55', 'R34', 'D71', 'R55', 'D58', 'R83']

test_wire1 = ['R98', 'U47', 'R26', 'D63', 'R33', 'U87', 'L62', 'D20', 'R33', 'U53', 'R51']
test_wire2 = ['U98', 'R91', 'D20', 'R16', 'D67', 'R40', 'U7', 'R15', 'U6', 'R7']
```

```python
def wire_path(wire_commands: list) -> list:
    path = []
    x, y = 0, 0
    for command in wire_commands:
        direction = command[0]
        mag = int(command[1:]) + 1
        if direction == 'R':
            for i in range(1, mag):
                x += 1
                path.append((x,y))
        elif direction == 'D':
            for i in range(1, mag):
                y -= 1
                path.append((x,y))
        elif direction == 'U':
            for i in range(1, mag):
                y += 1
                path.append((x,y))
        elif direction == 'L':
            for i in range(1, mag):
                x -= 1
                path.append((x,y))
    return path

def intersection(list1, list2):
    list3 = [x for x in list1 if x in list2]
    return list3

def min_manhat_dist(list1):
    return min([sum([abs(a) for a in x]) for x in list1])

test_path1 = wire_path(test_wire1)
test_path2 = wire_path(test_wire2)

intersect = intersection(test_path1, test_path2)
min_manhat_dist(intersect)
```

# Part One Solution

```python
path1 = wire_path(wire_1)
path2 = wire_path(wire_2)

crosses = intersection(path1, path2)
answer = min_manhat_dist(crosses)

print(f'The solution to part one is {answer}.')
```

# Part Two Solution

```python
# While solving part one, I created "path1" and "path2" which are lists that contain the ordered tuples that each wire moves
# through on it's journey. So the number of steps that a wire takes to get to any point is just it's index+1 in its respective
# path list. 

crosses = intersection(path1, path2)
min_steps_to_cross = len(path1) + len(path2) # Max number of steps that can be taken

for cord in crosses:
    steps1 = path1.index(cord) + 1
    steps2 = path2.index(cord) + 1
    min_steps_to_cross = min(min_steps_to_cross, steps1 + steps2)

print(f'The answer to part two is {min_steps_to_cross}.')
```
